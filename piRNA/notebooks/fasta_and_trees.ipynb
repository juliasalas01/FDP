```python
import os
from Bio import SeqIO

output_directory = "/home/juliasalas/piRNA/PrimaryData/primate.multiz_tarsier/fasta/"
os.makedirs(output_directory, exist_ok=True)

import pandas as pd

sequences = []

for maf_file in maf_files:
    alignment_iterator = AlignIO.parse(maf_file, "maf")
    alignment_iterator = list(alignment_iterator)

    dataparsed = pd.DataFrame()
    for block in alignment_iterator:
        ids = [s.id.split('.')[0] for s in block]
        dataparsed = pd.DataFrame(index=ids)
        dataparsed[0] = ''

        # concatenar
        for block in alignment_iterator:
            for s in block:
                dataparsed.loc[s.id.split('.')[0], 0] += str(s.seq)
            print(dataparsed)
   # Save all rows in a single FASTA file
    fasta_filename = maf_file.replace(".maf", ".fasta")
    fasta_file = os.path.join(output_directory, fasta_filename)
    with open(fasta_file, "w") as f:
        for row in dataparsed.iterrows():
            species = row[0]
            sequence = row[1][0]
            f.write(f">{species}\n")
            f.write(f"{sequence}\n")
        
```

# Parameters: identity and upgma

```python
os.chdir('/home/juliasalas/piRNA/PrimaryData/primate.multiz_tarsier/fasta/')
for file in fasta_files:
    print(file)
    # Read the alignment from file
    alignment = AlignIO.read(file, "fasta")
   

    # Calculate the distance matrix
    calculator = DistanceCalculator("identity")
    distance_matrix = calculator.get_distance(alignment)

    # Construct the tree using the neighbor-joining method
    constructor = DistanceTreeConstructor()
    tree = constructor.upgma(distance_matrix)

    # Print the file name
    print("Alignment:", file)

    # Print the ASCII tree
    Phylo.draw_ascii(tree)
    print()
```

"identity": Calculates the pairwise identity between sequences. It measures the percentage of identical positions in the alignment.

"blastn": Calculates the distance based on the alignment score obtained from the BLAST algorithm. This method is suitable for aligning divergent sequences.

"blosum62": Calculates the distance based on the BLOSUM62 substitution matrix. It considers the amino acid substitution probabilities from the BLOSUM62 matrix to estimate the distance.

"trans": Calculates the pairwise transition/transversion ratio between nucleotide sequences. This method is suitable for aligning nucleotide sequences.

"pam30": Calculates the distance using the method implemented in the PAML package (Phylogenetic Analysis by Maximum Likelihood). It estimates the distance based on the model of molecular evolution.
