```python
from Bio import Phylo, SeqIO
from Bio.Align import MultipleSeqAlignment
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor

fasta_file = "/home/juliasalas/piRNA/PrimaryData/primate.multiz_tarsier/chr22.filtered.region.37352189_2.fasta"



# Read the FASTA file
sequences = SeqIO.parse(fasta_file, "fasta")

# Create a list to store the sequence records
sequence_records = []

# Iterate over each sequence in the FASTA file
for sequence in sequences:
    sequence_records.append(sequence)

# Create distance calculator and tree constructor
calculator = DistanceCalculator("identity")
constructor = DistanceTreeConstructor(calculator, method="nj")

# Calculate distances and build the tree
alignment = MultipleSeqAlignment(sequence_records)
tree = constructor.build_tree(alignment)

Phylo.draw(tree, do_show=False)
```
