```r
library("dplyr")
library("ggplot2")
```

```r
setwd("~/piRNA/Workspaces/julia/PrimaryData/ref_data/orangutan")
```

```r
#chromosome data
chromosome_data<- read.table("summary_chroms.tsv", header=TRUE, sep="\t")
```

```r
setwd("~/piRNA/Workspaces/julia/piRNA_reads/mapping_species/coverage/Orang1")
```

```r
coverage<- read.table("Orang1_coverage", header = FALSE)
```

```r
head(coverage)
```

```r
names(coverage)[names(coverage) == "V1"]<- "REFSEQ_CHROM"
data<- merge(coverage, chromosome_data, by= "REFSEQ_CHROM")
```

```r
head(data)
table(data$UCSC_CHROM)
```

```r
x="Orang 1"
```

```r
#CHR1
chr1<- data %>% filter(UCSC_CHROM == "chr1")
```

```r
ggplot(chr1, aes(V2, V3)) +
  geom_line(color="#56B4E9") +
  theme_minimal()+
  ggtitle(paste0(x," coverage chr1")) +
    labs(x="Position", y="Coverage")
```
